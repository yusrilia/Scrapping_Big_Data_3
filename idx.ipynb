### Latar Belakang

Dalam era **Big Data**, pengolahan data keuangan menjadi krusial untuk analisis bisnis dan pengambilan keputusan. Bursa Efek Indonesia (IDX) menyediakan laporan keuangan perusahaan dalam format **XBRL/XML**, yang kaya akan informasi tetapi memerlukan proses ingestion agar dapat digunakan secara efektif. Data ini harus diunduh, diurai dengan parser XML, lalu disimpan ke dalam database yang mendukung struktur data kompleks, seperti **MongoDB**. Proses ini memastikan data dapat diakses, dianalisis, dan diintegrasikan dengan sumber lain untuk menghasilkan wawasan yang bernilai.

```python
import zipfile
import os
import xml.etree.ElementTree as ET
import json

# Folder tempat ZIP berada (folder kerja saat ini)
zip_folder = os.getcwd()  # Mengambil folder kerja saat ini
extract_folder = os.path.join(zip_folder, "extracted")  # Folder untuk mengekstrak file
json_output_file = os.path.join(zip_folder, "combined_data.json")  # File JSON hasil gabungan

# Pastikan folder ekstraksi ada
if not os.path.exists(extract_folder):
    os.makedirs(extract_folder)

# Daftar untuk menyimpan semua data
all_data = []

# Ekstrak semua file ZIP yang ada di dalam folder zip_folder
for zip_file in os.listdir(zip_folder):
    if zip_file.endswith(".zip"):  # Cek apakah file tersebut berformat .zip
        zip_path = os.path.join(zip_folder, zip_file)
        company_name = zip_file.split(".")[0]  # Misal: "ASII"
        
        # Buat folder khusus untuk ekstraksi perusahaan
        company_extract_folder = os.path.join(extract_folder, company_name)
        if not os.path.exists(company_extract_folder):
            os.makedirs(company_extract_folder)

        # Ekstrak file ZIP ke folder perusahaan
        with zipfile.ZipFile(zip_path, "r") as zip_ref:
            zip_ref.extractall(company_extract_folder)

        print(f"Ekstrak selesai: {zip_file} ke {company_extract_folder}")

# Loop semua file XBRL dalam folder hasil ekstraksi
for company in os.listdir(extract_folder):  # Loop untuk setiap folder perusahaan
    company_path = os.path.join(extract_folder, company)
    
    if os.path.isdir(company_path):  # Pastikan ini adalah folder perusahaan
        for file in os.listdir(company_path):  # Loop file dalam folder perusahaan
            if file.endswith(".xbrl"):  # Pastikan hanya file XBRL yang diproses
                xbrl_path = os.path.join(company_path, file)
                
                try:
                    # Parsing XML/XBRL
                    tree = ET.parse(xbrl_path)
                    root = tree.getroot()

                    # Simpan data dalam dictionary
                    data = {"filename": file, "company": company}
                    for elem in root.iter():
                        tag = elem.tag.split("}")[-1]  # Hapus namespace dari tag XML
                        data[tag] = elem.text  # Simpan teks dari elemen

                    # Tambahkan data perusahaan ke dalam list all_data
                    all_data.append(data)

                    print(f"Data dari {file} untuk perusahaan {company} berhasil ditambahkan.")

                except Exception as e:
                    print(f"Gagal memproses {file}: {e}")

# Simpan semua data dalam satu file JSON
try:
    with open(json_output_file, "w", encoding="utf-8") as json_file:
        json.dump(all_data, json_file, indent=4)

    print(f"Semua data berhasil disimpan dalam {json_output_file}")
except Exception as e:
    print(f"Gagal menyimpan data ke file JSON: {e}")
```

### Penyimpanan data Ingestion ke MongoDB

```python
import json
from pymongo import MongoClient

# Fungsi untuk memasukkan data JSON ke MongoDB
def insert_json_to_mongodb(json_file, db_name, collection_name):
    # Connect ke MongoDB (ganti dengan URL MongoDB lokal atau dari MongoDB Atlas jika perlu)
    client = MongoClient("mongodb://localhost:27017/")  # URL MongoDB lokal
    db = client[db_name]  # Pilih database
    collection = db[collection_name]  # Pilih koleksi
    
    # Baca file JSON
    with open(json_file, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    # Masukkan data ke MongoDB
    if isinstance(data, list):
        # Jika data adalah list (seperti hasil gabungan XBRL)
        collection.insert_many(data)
    else:
        # Jika hanya satu objek (misalnya satu dokumen)
        collection.insert_one(data)

    print(f"Data berhasil dimasukkan ke MongoDB dalam koleksi '{collection_name}'.")

# Path ke file JSON yang ingin dimasukkan ke MongoDB
json_file = "combined_data.json"  # Ganti dengan path file JSON yang sesuai

# Nama database dan koleksi di MongoDB
db_name = "financial_reports"  # Ganti nama database sesuai kebutuhan
collection_name = "reports_data"  # Ganti nama koleksi sesuai kebutuhan

# Masukkan data ke MongoDB
insert_json_to_mongodb(json_file, db_name, collection_name)
```
